### Contents of paper
#### How it works?
##### Approach to art

First of all, it starts with explaining how to approach to ‘Art’ which is separation of content and style. When it comes to art generated by human, they say It can be explained by complex interplay between content and style. Same in neural model, these two factors are extracted with pre-trained CNN model and regulated by weights and loss functions so that we can make white noise image, initial y, art. Let’s start from beginning again. We need pre-trained CNN and content image that you want to draw and style image that you want to copy its style. In this paper, they use VGG model that have 19 layers. 

##### Content

Let’s talk about content first. What we need to know is principal of CNN. When image file enters into CNN model, first few layers catch minor features like edge, line, and motifs. The deeper layers are, the more complex its features that they catch like object. 
What if we put content image file into CNN? Obviously, its features are caught by each layer. For content image, we are supposed to extract its content as global arrange not as local like motifs. Hence, we are to choose one layer [l] which is not too shallow and not too deep. Next, we put x which is white noise image (randomized image) into CNN model and pick same layer [l] for extracting its content. We can say P and F as respective feature value of content image and x. Square-error loss between P and F is content loss function which is one of two loss functions in this paper. 

##### Style

Method to extract style representation is quite similar to content’s one. In this part, Gram matrix is introduced to express correlation of its specific layer. Again, Gram matrix is defined correlation for all channels in layer [l]. We can get gram matrix A from style image and G from x, which is also white noise image. Then, E is defined by mean squared error between A and G. And style loss function is defined by E with w, weights for each layer, from [l]=0 to [l]=L. Finally, total loss function is defined by alpha times content loss function plus beta times style loss function. Alpha and beta is regularization factor to control two loss functions. Through this, we can make fancy art with neural networks.
